## What is a dictionary

A dictionary is a data structure that stores data objects that linked to keys. Data objects can be numbers (integers or floats), string ("text in quotations"), data frames, etc...

What are keys?: Keys are inputs that link to a particular output.

# More Notes on Dictionaries: Fundamentals

As a reminder, dictionaries have keys, words, that link to objects, or definitions. Dictionaries use curly brackets. The form for dictionaries is {key:obj}

```python
dct = {}
dct = {"listen": "to hear and take in information, music, other sounds",
      "speak": "to make sounds outloud to convey meaning or lack of"}
dct
```

To include multiple definitions for each word you need to create a new dictionary for each entry.

```python
dct = {"listen": {},
      "speak": {}}

dct
```

```python
dct = {"dry": {"verb": "the act of making an object not wet",
                  "adjective": "the state of an object when not wet"},
      "clean": {"verb": "the act of purifying a space, body, or other object",
               "adjective": "the state of an object when tidy, pure, sanitary"}}

dct
```

The second layer of keys is the same (verb and adjective - {"verb" : def, "adjective" : def}). So, a dataframe (pandas as pd) will automatically make (otherwise known as index) the rows for us.

```python
import pandas as pd
df = pd.DataFrame(dct)
df
```
